```python
y = furniture['Sales'].resample('MS').mean()
print(y['2017':])
```

```python
#DATA VISUALIZATION
y.plot(figsize=(15,6))
plt.show()
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 18,8
decomposition = sm.tsa.seasonal_decompose(y,model= 'addictive')
fig = decomposition.plot()
plt.show()
```
